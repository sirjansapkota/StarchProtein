```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import distance
from scipy.cluster import hierarchy

%matplotlib inline
os.chdir('/panicle/ssapkot/git_repo/StarchProtein/')
```

## Significant SNPs

```python
df = pd.read_csv('results/expression/RIKEN_GeneExp_sigGenes_range.csv',index_col=0)
print(df.head())
g = sns.clustermap(df,cmap='RdYlBu')
g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 12, rotation=45)
g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_ymajorticklabels(), fontsize = 10)
plt.savefig('results/expression/Heatmap_sigGenes_Expression_RIKEN.png', bbox_inches='tight', dpi=500)
```
