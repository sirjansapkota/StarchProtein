### SAP Grain Composition Per Grain values to get BLUPs for each Year

```r
main <- read.csv('data/SAP_GrainComp_PerGrain_Complete.csv',header=T)
head(main)
main$Year <- as.factor(main$Year)
main$Rep <- as.factor(main$Rep)
str(main)
```

```r
df <- main[main$Year==2017,]
taxa <- read.csv('../../GrainComp/data/pheno.csv')
BLUPs <- as.data.frame(taxa[,1])
traits <- colnames(df[,-c(1:4)])
attach(df)
############ for getting BLUPs
#BLUPs <- as.data.frame(unique(SC2014$Taxa))
colnames(BLUPs) <- "Taxa"
for (i in 1:length(traits)){
  trait <- traits[i]
  trait_df <- df[,trait]
  fm <- lmer(trait_df ~ (1|Taxa) + (1|Rep))
  rr <- ranef(fm)
  blup <- as.data.frame(rr$Taxa)
  taxa <- rownames(blup)
  trt <- paste0(trait,"2017")
  blup_df <- as.data.frame(cbind(taxa,blup$`(Intercept)`))
  colnames(blup_df) <- c("Taxa",trt)
  BLUPs <- BLUPs %>% left_join(blup_df, by = "Taxa")
}
head(BLUPs)
#BLUPs_all <- BLUPs
detach(df)
```

```r

#BLUPs_all <- BLUPs_all %>% left_join(BLUPs, by = "Taxa")
str(BLUPs_all)
pheno <- as.data.frame(lapply(BLUPs_all, function(x) as.numeric(levels(x))[x]))
phenoBME <- cbind(BLUPs_all$Taxa,pheno[,-1])
colnames(phenoBME) <- colnames(pheno)
head(phenoBME)    
str(phenoBME)
write.csv(phenoBME, 'data/SAP_perGrainComp_BME.csv',row.names=FALSE)
#############
```
